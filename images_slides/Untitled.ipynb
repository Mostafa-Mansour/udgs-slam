```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
depth_img = plt.imread("neural_depth.jpg")
```

```python
pixel_intensities = depth_img[:,:,0].flatten()/100

# Create a box plot of the pixel intensities
plt.figure(figsize=(8, 6))
plt.boxplot(pixel_intensities)
plt.title('Box Plot of depth values')
plt.ylabel('Depthes (m)')
plt.xticks([1], ['Depth image'])
plt.grid(True)
plt.savefig('neural_depth.png', bbox_inches='tight')
```

```python
depth_img_filtered = plt.imread("neural_depth_filtered.png")
pixel_intensities = depth_img_filtered[:,:,0].flatten()

# Create a box plot of the pixel intensities
plt.figure(figsize=(8, 6))
plt.boxplot(pixel_intensities)
plt.title('Box Plot of depth values')
plt.ylabel('Depthes (m)')
plt.xticks([1], ['Depth image'])
plt.grid(True)
plt.savefig('neural_depth.png', bbox_inches='tight')
```
